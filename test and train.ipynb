```python
import numpy as np 
import pandas as pd 
```

```python
orders = pd.read_csv('orders.csv')
orders
```

```python

grouped=orders.groupby(['user_id'])["order_number"].aggregate("max").reset_index()
grouped["eval_set"]="TT"
grouped
```

```python
prior1 = pd.merge(orders,grouped, on = ["user_id","order_number"],  how = "left")
prior1
```

```python
prior2=prior1.fillna(value='Prior')
prior2
```

```python
prior = prior2[prior2.eval_set == "Prior"]
prior
```

```python
orders.loc[:,'order_number'].value_counts()
```

```python
order_products = pd.read_csv('order_products.csv')
order_products
```

```python
reorder = pd.merge(orders,order_products, on = "order_id",  how = "left")
```

```python
reorder
```

```python
test = reorder[reorder['product_id'].isnull()]
test
```

```python
test['eval_set']='Test'
test
```

```python
test_new=test.iloc[:,[0,1,2,3,4,5,9]]
test_new
```

```python
test_new.order_id.nunique()
```

```python
frames=[test_new,prior]
result = pd.concat(frames)
result
```

```python
result_new=result.iloc[:,[0,1,2,3,4,5,6]]
result_new
```
