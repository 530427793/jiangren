```python
import numpy as np 
import pandas as pd 
```

```python
orders = pd.read_csv('orders.csv')
orders
```

```python
order_products = pd.read_csv('order_products.csv')
order_products
```

```python
reorder = pd.merge(order_products, orders, on = "order_id",  how = "left")
reorder
```

```python
train = orders.groupby(['user_id'])["order_number"].aggregate("max").reset_index()
train
```
