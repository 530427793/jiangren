```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
```

```python
order_products = pd.read_csv('order_products.csv')
```

```python
orders = pd.read_csv('orders.csv')
```

```python
reorder = pd.merge(order_products, orders, on = "order_id",  how = "left")
```

```python
reorder.head(30)
```

```python
reorder.shape
```

```python
reorder1=reorder.groupby(['user_id', 'product_id'])["order_id"].aggregate("min").reset_index()
```

```python
reorder1
```

```python
reorder1["reorder"]=0
```

```python
reorder1.shape
```

```python
reorder2 = pd.concat([order_products, reorder1], axis=1, join='outer')
reorder2
```

```python
reorder2.shape
```

```python
reorder2['reorder']=reorder2['reorder'].fillna(1)
```
